# <font color = "blue">Pie Chart.</font>

- When the data is having Categorical we can use this Pie Chart

- we can check **Comparision and Composition**

- **Univariate/Bivariate Analysis**
- **Categorical vs numerical**

**Syntax**

matplotlib.pyplot.pie(data, explode=None, labels=None, colors=None, autopct=None, shadow=False)

- **Data**       == This Parameter Specifies that Data u want to display in Graph.
- **Labels**     == This Parameter Specifies that Data u want to display in Graph.
- **Colors**     == Adding manual colors.
- **autopct**    == Is a string used to label the wedge with their numerical value.
- **Shadow**     == Is used to create shadow of wedge. 
- **Explode**    == Separting Slices

###  <font color = "magenta">                  </font> 

###  <font color = "magenta"> Basic pie Graph.                  </font>

```python
# Creating sample Data


import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

marks = [20,30,40,50,60]
sub   = ["Eng","Phy","Maths","Social","Chemistry"]


plt.pie(marks,labels=sub)
plt.show()
```

### <font color = "magenta">                 </font>

### <font color = "magenta">Creating DataFrame and plotting Pie Chart.                 </font>

```python
df = pd.DataFrame({"Batsman":["Dhoni","Gaikwad","Conway","Jaddu"],
                   "Who Scored more runs":[20,36,70,26]})
df


plt.pie(df["Who Scored more runs"],labels = df["Batsman"])
plt.show()
```

### <font color = "magenta">        </font>

### <font color = "magenta">Adding Percentages.        </font>

```python
plt.pie(df["Who Scored more runs"], labels = df["Batsman"], autopct = '%0.1f%%')
plt.show()
```

### <font color = "magenta">         </font> 

### <font color = "magenta">Adding Colors.         </font> 

```python
plt.pie(df["Who Scored more runs"], labels = df["Batsman"], autopct = '%0.1f%%',
       colors = ["Yellow","orange","purple","brown"])
plt.show()
```

### <font color = "magenta">                </font>

### <font color = "magenta">Explode Shadow.                </font>

```python
plt.pie(df["Who Scored more runs"], labels = df["Batsman"], autopct = '%0.1f%%',
       colors = ["Yellow","orange","purple","brown"],explode=[0.2,0.1,0,0.3],shadow=True)
plt.show()
```
