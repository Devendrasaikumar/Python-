# <font color = "blue">Histogram</font>

- When the data is having Categorical we can use this Histogram.
- To check **Comparison** we can use this one

- **Univariate Analysis**
- **Numerical col**
- **Use case - Frequency Count**

**Syntax**

 - matplotlib.pyplot.hist(x, bins, density, color, label, edgecolor, kde)
 
 - **X**            == Data.
 - **bins**         == optional parameter contains integer or sequence or strings.
 - **Density**      == Optional parameter contains boolean values.
 - **Color**        == optional parameter used to set color or sequence of color specs.
 - **label**        == optional parameter string or sequence of string to match with multiple datasets.
 - **Edgecolor**    == Baredge color fillup
 - **kde**          == Checking Density 

### <font color = "magenta">                </font>

### <font color = "magenta">Basic Histogram                </font>

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate random data for the histogram
np.random.seed(456)
data = np.random.randn(1000)
data

plt.hist(data, bins=30)

plt.show()
```

### <font color = "magenta">                </font>

### <font color = "magenta">Adding color and its border color to Histogram                </font>

```python
data = np.random.randn(1000)
data

plt.hist(data, bins=30,color = "red", edgecolor = "white")

plt.show()
```

### <font color = "Magneta">          </font>

### <font color = "Magenta">Plotting Two Histograms at single frame         </font>

```python
data1 = np.random.randn(1000)
data1

data2 = np.random.normal(loc=3, scale=1, size=1000)
data2


# Location setting
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 4))



# Histogram Data1
axes[0].hist(data1, bins=30,color = "red", edgecolor = "black")
axes[0].set_title("Histogram1")


# Histogram Data2
axes[1].hist(data2, bins=30,color = "gold", edgecolor = "black")
axes[1].set_title("Histogram2")

plt.show()
```

### <font color = "Magenta">               </font>

### <font color = "Magenta">Stacked Histogram.               </font>

```python
# Creating a stacked histogram
plt.hist([data1, data2], bins=30, stacked=True, color=['red', 'gold'], edgecolor='black')
plt.show()
```

### <font color = "Magenta">                    </font>

### <font color = "magenta">Density Plot                </font>

```python
import seaborn as sns
import matplotlib.pyplot as plt

data = np.random.randn(1000)
data

sns.distplot(data, bins=30, kde=True, color='red', label = "Values")

plt.title("Checking Density")
plt.xlabel("Data Values")
plt.ylabel("Frequency")

plt.legend()
plt.grid()
plt.show()
```
