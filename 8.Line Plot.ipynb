# <font color='Blue'>Line Graph.</font>


- When the data is continous we can use this line Graph.
- To check **Flow, Relationship, Comparsion, and Trend**

- **Bivariate Analysis**
- **categorical -> numerical and numerical -> numerical**
- **Use case - Time series data**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt



# Plotting simple graph using numpy array data


Price = np.array([450,470,500,530,580,600])
Year =  np.array([2015,2016,2017,2018,2019,2020])

plt.plot(Price,Year)
plt.show()
```

- **Now we can create new dataframe and will try**

```python
import pandas as pd
import matplotlib.pyplot as plt


df=pd.DataFrame({"Year":[2010,2011,2012,2013,2014,2015],
                 "Dhoni":[400,600,480,400,410,300],
                 "kohli":[200,350,500,560,600,620]})
df
```

##### <font color='magenta'>Extracting from DataFrame to draw a single Line Graph.</font>


- Extracting from above DataFrame and plotting Dhoni runs how many scored Each Year


```python
plt.plot(df["Year"],df["Dhoni"])
plt.show()
```

##### <font color='brown'>                               .</font>


##### <font color='magenta'>Multiple Line Graph.</font>


- Plotting multiple lines  And check Comparison runs to each other 


```python
plt.plot(df["Year"],df["Dhoni"])
plt.plot(df["Year"],df["kohli"])
plt.show()
```

##### <font color='brown'>                               .</font>


##### <font color='magenta'>Adding titles (Labels) to Line Graph.</font>


- Adding Titles to  above graph. By Using

- {    **plt.title()---->Main Title**

- **plt.xtitle()--->X-Axis**

- **plt.ytitle()--->Y-Axis**      }


```python
plt.plot(df["Year"],df["Dhoni"])
plt.plot(df["Year"],df["kohli"])

plt.title("Dhoni Vs Kohli Runs Difference")       # Main Title

plt.xlabel("Year")                               # X-Axis Title 

plt.ylabel("Runs")                               # Y-Axis Title

plt.show()
```

##### <font color='brown'>                               .</font>


##### <font color='magenta'>Adding Colors to Line Graph.</font>


- By defualt, it takes Orange And Blue. We Can change the color of Lines by using ( **color = "  "** )


```python
plt.plot(df["Year"],df["Dhoni"], color="cyan")
plt.plot(df["Year"],df["kohli"], color="purple")

plt.title("Dhoni Vs Kohli Runs Difference")

plt.xlabel("Year")

plt.ylabel("Runs")

plt.show()
```

##### <font color='brown'>                         </font>


##### <font color='magenta'>Adding Dotted Line and Dashed Line to Line Graph.</font>

- We are adding Linetype Dotted are Dashed Line type. By Using ( **linestyle = " "** )


```python
plt.plot(df["Year"],df["Dhoni"], color="cyan", linestyle = "dotted")
plt.plot(df["Year"],df["kohli"], color="purple", linestyle = "dashed")

plt.title("Dhoni Vs Kohli Runs Difference")

plt.xlabel("Year")

plt.ylabel("Runs")

plt.show()
```

##### <font color='brown'>                         </font>

##### <font color='magenta'> Incressing Line Width in Line Graph.</font>

- we can incress linewidth. By using ( **linewidth = number**)

```python
plt.plot(df["Year"],df["Dhoni"], color="cyan", linestyle = "dotted", linewidth = 5)
plt.plot(df["Year"],df["kohli"], color="purple", linestyle = "dashed", linewidth = 3)

plt.title("Dhoni Vs Kohli Runs Difference")

plt.xlabel("Year")

plt.ylabel("Runs")

plt.show()
```

##### <font color='brown'>                         </font>

##### <font color='magenta'> Marking points and its Size in Line Graph.</font>

- we can keep marking point. By using ( **marker = "shapename"** )
- and also we can incress marking size ( **markersize = number** )

```python
plt.plot(df["Year"],df["Dhoni"], color="cyan", linestyle = "dotted", linewidth = 5, marker = "x", markersize = 10)
plt.plot(df["Year"],df["kohli"], color="purple", linestyle = "dashed", linewidth = 3,marker = "*", markersize = 15)

plt.title("Dhoni Vs Kohli Runs Difference")

plt.xlabel("Year")

plt.ylabel("Runs")

plt.show()
```

##### <font color='brown'>                         </font>

##### <font color='magenta'> Kepping Legend  in Line Graph AND its Location.</font>

- We Can keep Lengend symbol in graph. By using ( **label = "Name"** )

- And at the end use **plt.legend()**

- For, Location use **plt.legend( loc= "upper Right" )** 



```python
plt.plot(df["Year"],df["Dhoni"], color="cyan", linestyle = "dotted", linewidth = 5, marker = "x", markersize = 10, label="Dhoni")
plt.plot(df["Year"],df["kohli"], color="purple", linestyle = "dashed", linewidth = 3,marker = "*", markersize = 15, label="Kohli")

plt.title("Dhoni Vs Kohli Runs Difference")
plt.xlabel("Year")
plt.ylabel("Runs")

plt.legend(loc = "center right")
plt.show()
```

##### <font color='brown'>                         </font>

##### <font color='magenta'> Limiting axes.</font>

- we can kepp limit to the X-axis. By using ( **plt.xlim()** )

- - we can kepp limit to the Y-axis. By using ( **plt.ylim()** )


```python
plt.plot(df["Year"],df["Dhoni"], color="cyan", linestyle = "dotted", linewidth = 5, marker = "x", markersize = 10, label="Dhoni")
plt.plot(df["Year"],df["kohli"], color="purple", linestyle = "dashed", linewidth = 3,marker = "*", markersize = 15, label="Kohli")

plt.title("Dhoni Vs Kohli Runs Difference")
plt.xlabel("Year")
plt.ylabel("Runs")

plt.xlim(2010,2014)
plt.ylim(0,1000)

plt.legend(loc = "center right")
plt.show()
```

##### <font color='brown'>                         </font>

##### <font color='magenta'> Grid.</font>

- Using grid() background it will come lines


```python
plt.plot(df["Year"],df["Dhoni"], color="cyan", linestyle = "dotted", linewidth = 5, marker = "x", markersize = 10, label="Dhoni")
plt.plot(df["Year"],df["kohli"], color="purple", linestyle = "dashed", linewidth = 3,marker = "*", markersize = 15, label="Kohli")

plt.title("Dhoni Vs Kohli Runs Difference")
plt.xlabel("Year")
plt.ylabel("Runs")

plt.legend(loc = "center right")
plt.grid()
plt.show()
```
